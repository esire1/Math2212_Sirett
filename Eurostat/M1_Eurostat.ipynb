# Adding rows to a dataframe

The idea is to make the collection of new rows into a new database. Then you just 'glue' that new database onto the end of your original database.

```python
# to create a new database, start with a dictionary
new_rows_dict = {
    'TIME' : [2000, 2000, 2000],
    'GEO' : ['USA', 'Canada', 'Uganda'],
    'Value' : [5, 4.18, 5.15]
}

# turns the dictionary into a dataframe
new_rows_df = pd.DataFrame(new_rows_dict)

new_rows_df
```

```python
# append the new_rows_df to the end of edu
edu_new_rows = edu.append(new_rows_df, ignore_index=True, sort=False)

edu_new_rows.tail()
```
