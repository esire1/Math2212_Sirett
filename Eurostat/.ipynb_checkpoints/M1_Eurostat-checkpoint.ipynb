# Adding rows to a dataframe

The idea is to make the collection of new rows into a new database. Then you just 'glue' that new database onto the end of your original database.

```python
# to create a new database, start with a dictionary
new_rows_dict = {
    'TIME' : [2000, 2000, 2000],
    'GEO' : ['USA', 'Canada', 'Uganda'],
    'Value' : [5, 4.18, 5.15]
}

# turns the dictionary into a dataframe
new_rows_df = pd.DataFrame(new_rows_dict)

new_rows_df
```

```python
# append the new_rows_df to the end of edu
edu_new_rows = edu.append(new_rows_df, ignore_index=True, sort=False)

edu_new_rows.tail()
```

```python
# worksheet 4
Worksheet4_dict = {
    'TIME' : [2001, 2001],
    'GEO' : ['Alaska', 'Hawaii'],
    'Value' : [.2, .3]
}
Worksheet4_df = pd.DataFrame(Worksheet4_dict)

edu_new_rows = edu.append(new_rows_df, ignore_index=True, sort=False)

edu_new_rows.head()
```

## Aggregating Data
sum or average or otherwise aggregate data that shares a similar property

```python
edu_ws4 = edu[['TIME','GEO']].groupby('TIME').mean()
edu_ws4

# error bc you can't take the mean of GEO
# worksheet 4 -- find the error
```

```python
# worksheet 5 -- print a table that contains the product of all 'Vales' within each year
edu_ws5 = edu [['TIME','Value']].groupby(['TIME']).prod()
edu_ws5
```

## Pivot Tables

A pivot table is similar to the tables using .groupby() but instead of aggregating over one variable, it aggregates over two variables. 

```python
# pd pulls from the pandas library
ptable = pd.pivot_table(
# database
    edu,
# one we are aggregating
    values = 'Value',
# what/rows we are grouping by
    index = 'TIME',
#what/columns we are grouping by
    columns = 'GEO',
# aggregating function, do not need parenthesis
    aggfunc = np.sum
)
ptable
```

```python
# worksheet 6, pivot table, countries for rows, year for columns, mean value
edu_ws6 = pd.pivot_table(
    edu,
    values = 'Value',
    index = 'GEO',
    columns = 'TIME',
    aggfunc = np.mean
)
edu_ws6
```

## Plotting Data

Another way of presenting data and making it more presentable.

The basic format to plot data is the same (see notes)

```python
# scatter plot
edu.plot(x= 'TIME', y = 'Value', kind = 'scatter', title = 'edu')
```

```python
# line plot for one country
edu_austria = edu[edu['GEO'] == 'Austria']
edu_austria.plot(x= 'TIME', y = 'Value', kind = 'line', title = 'edu austria')
```

```python
# line for two countries
edu_austria_france = edu[(edu['GEO'] == 'Austria') | (edu['GEO']== 'France')]
edu_austria_france.plot(x= 'TIME', y = 'Value', kind = 'line', title = 'edu austria and france')
```

```python
# bar graphs
# a bar graph of the total value within each region
edu_country_value_aggregate = edu [['GEO','Value']].groupby('GEO').sum()

edu_country_value_aggregate.plot(kind = 'bar')
```

```python
# bar graphs
# a bar graph of the total value within each region
# flips it
edu_country_value_aggregate = edu [['GEO','Value']].groupby('GEO').sum()

edu_country_value_aggregate.plot(kind = 'barh')
```
