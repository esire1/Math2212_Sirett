# Question 4
Create two new dataframes containing all columns of the original 'census' dataframe restrictied to include: 

Only males with the "total family income" in increasing order.  

Only females with the "total family income" in decreasing order. 

```python
cen_male= cen[cen['sex']=='Male']
cen_male.sort_values(by='total_family_income', ascending=True, inplace=True)
cen_male
```

```python
cen_female= cen[cen['sex']=='Female']
cen_female.sort_values(by='total_family_income', ascending=False, inplace=True)
cen_female
```

# Question 5
Create and print a new dataframe that has the average "total family income" in each state.

```python
cen_hw5 = cen [['state_fips_code','total_family_income']].groupby('state_fips_code').mean()
cen_hw5
```

# Question 6
Create and print a new dataframe with the average age for each race.

```python
cen_hw6 = cen[['race_general','age']].groupby('race_general').mean()
cen_hw6
```

# Question 7
Print a pivot table that gives the average personal income in each state, separating by sex. 

```python

cen_hw7 = pd.pivot_table(
    cen,
    values = 'total_personal_income',
    index = 'state_fips_code',
    columns = 'sex',
    aggfunc = np.mean
)
cen_hw7
```
