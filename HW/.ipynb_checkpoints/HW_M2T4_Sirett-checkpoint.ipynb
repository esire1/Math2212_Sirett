## Question 1
Find means, standard deviations, medians, and interquartile ranges for:

A. Annual income

B. Debt-to-income ratio

C. The difference between the total credit available and credit used

D. Interest rate

#### A. Annual income

```python
print loan['annual_income'].mean()
print loan['annual_income'].std()
print loan['annual_income'].median()
def IQR(x):
    return x.quantile(.75) - x.quantile(.25)

print IQR(loan['annual_income'])
```

#### B. Debt-to-income ratio

```python
print loan['debt_to_income'].mean()
print loan['debt_to_income'].std()
print loan['debt_to_income'].median()
def IQR(x):
    return x.quantile(.75) - x.quantile(.25)

print IQR(loan['debt_to_income'])
```

#### C. The difference between the total credit available and credit used

```python
difference = loan['balance'] - loan['paid_total']
```

```python
print difference.mean()
print difference.std()
print difference.median()
def IQR(x):
    return x.quantile(.75) - x.quantile(.25)

print IQR(difference)
```

#### D. Interest Rate

```python
print loan['interest_rate'].mean()
print loan['interest_rate'].std()
print loan['interest_rate'].median()
def IQR(x):
    return x.quantile(.75) - x.quantile(.25)

print IQR(loan['interest_rate'])
```

## Question 5

Remove some outliers from the debt-to-income ratio using the mean method and again using the median methods. Draw boxplots before and after removing the outliers.

```python
loan.boxplot('debt_to_income')
```

```python
debt = loan['debt_to_income']
#income = adult['income']
debt_median = debt.median()
debt_iqr = debt.quantile(0.75) - debt.quantile(0.25)

print len(loan[debt> debt_median+1.5*debt_iqr])
print len(loan[debt< debt_median-1.5*debt_iqr])

#print len(loan[(loan['income'] == '>50K')&(debt> debt_median+1.5*debt_iqr)])
#print len(loan[(loan['income'] == '>50K')&(debt<debt_median-1.5*debt_iqr)])

loan_median_outlier_drop = loan.drop(
    loan.index[
        (debt> debt_median+1.5*debt_iqr) | (debt<debt_median-1.5*debt_iqr)
    ]
)

print debt_median - 1.5*debt_iqr
print debt_median + 1.5*debt_iqr

loan_median_outlier_drop.head()
```

```python
loan_median_outlier_drop.boxplot('debt_to_income')
```

```python
debt = loan['debt_to_income']
#income = adult['income']
debt_mean = debt.mean()
debt_iqr = debt.quantile(0.75) - debt.quantile(0.25)

print len(loan[debt> debt_mean+1.5*debt_iqr])
print len(loan[debt< debt_mean-1.5*debt_iqr])

#print len(loan[(loan['income'] == '>50K')&(debt> debt_median+1.5*debt_iqr)])
#print len(loan[(loan['income'] == '>50K')&(debt<debt_median-1.5*debt_iqr)])

loan_mean_outlier_drop = loan.drop(
    loan.index[
        (debt> debt_mean+1.5*debt_iqr) | (debt<debt_mean-1.5*debt_iqr)
    ]
)

print debt_mean - 1.5*debt_iqr
print debt_mean + 1.5*debt_iqr

loan_mean_outlier_drop.head()
```
