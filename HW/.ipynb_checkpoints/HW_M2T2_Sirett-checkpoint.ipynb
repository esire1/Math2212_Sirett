```python
import  pandas  as pd
import  numpy  as np
import  matplotlib.pyplot  as plt
```

```python
cen = pd.read_csv('census.csv')
cen
```

```python
cen_marital_status = cen [['marital_status','total_family_income']].groupby('marital_status').mean()

cen_marital_status.plot(kind = 'bar')
```

Categories are divorced, married/spouse absent, married/spouse present, never married/single, separated, and widowed
