```python
import  pandas  as pd
import  numpy  as np
import  matplotlib.pyplot  as plt
```

```python
loan = pd.read_csv('loans_full.csv')
loan
```

## Question 1

```python
loan_corr_matrix = loan.corr()
print loan_corr_matrix
```

```python
import seaborn as sns

sns.heatmap(
    loan_corr_matrix,
    xticklabels = loan_corr_matrix.columns,
    yticklabels = loan_corr_matrix.columns,
    annot=True
)
```

## Question 2

```python
column_annual_income = loan['annual_income']
column_debt_to_income = loan['debt_to_income']
correlation_A = column_annual_income.corr(column_debt_to_income)
print correlation_A
```

```python
column_annual_income = loan['annual_income']
column_interest_rate = loan['interest_rate']
correlation_B = column_annual_income.corr(column_interest_rate)
print correlation_B
```

```python
column_debt_to_income = loan['debt_to_income']
column_interest_rate = loan['interest_rate']
correlation_C = column_debt_to_income.corr(column_interest_rate)
print correlation_C
```
